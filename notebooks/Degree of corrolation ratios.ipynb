# Degrees of corrolation in rectilinear, diagonal and circular basis versus cross dephasing lifetime.
No fine structure, no secondary emission, no background emission, no spin scattering.

```python
rootdir = '../out/2013-08-03/crosstau/'
dat1 = get_data(rootdir, 'crosstau')
plot_deg_of_corr(dat1)
```

```python

hold_param1 = dat1[0].astype(np.float)
grect1 = dat1[1][0]
gdiag1 = dat1[1][1]
gcirc1 = dat1[1][2]

plt.figure(figsize = (16/1.5, 9/1.5))
plt.plot(hold_param1, gdiag1/ grect1, 'r--')
plt.plot(hold_param1, - gcirc1/ grect1, 'b--')

plt.xlabel('Background emission rate.') ;
plt.legend(['gdiag/grect', '-gcirc/grect'])
```

# Degree of corr ratio with 1.5$\mu eV$ FSS,  2.5ns decoherence time and 0.0 -> 0.5 background emission rate

```python
rootdir = '../out/2013-08-06/fid_v_bg/'
dat = get_data(rootdir, 'bg_emission_rate')
plot_deg_of_corr(dat)
plt.gca().invert_xaxis()
plt.xlabel('Background emission rate')
```

```python
hold_param = dat[0].astype(np.float)
grect = dat[1][0]
gdiag = dat[1][1]
gcirc = dat[1][2]
from scipy import stats
slope1, intercept1, a, b, c= scipy.stats.linregress(hold_param, gdiag/grect)
slope2, intercept2, a, b, c= scipy.stats.linregress(hold_param,  - gcirc/grect)
```

```python

plt.figure(figsize = (16/1.5, 9/1.5))
plt.gca().invert_xaxis()
plt.plot(hold_param, gdiag/ grect, 'r--', hold_param, intercept1 + slope1*hold_param, 'r-')
plt.plot(hold_param, - gcirc/ grect, 'b--', hold_param, intercept2 + slope2*hold_param, 'b-')

plt.xlabel('Background emission rate.') ;
plt.legend(['gdiag/grect', 'fit', '-gcirc/grect', 'fit'])
plt.ylim([0, 0.5])
```

In the reasonably realistic background range of 0 -> 0.3 the ratios of grect to gdiag and gcirc looks fairly constant.
