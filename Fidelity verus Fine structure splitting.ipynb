```python
%pylab inline
from mpl_toolkits.axes_grid1 import AxesGrid
from Icarus.Classes.QuantumDot import QuantumDot
from constants import Constants
```

# Fidelity versus Fine Structure splitting.

```python
constants = Constants()
qd = QuantumDot(constants.xtau, constants.xxtau, constants.ptau, constants.FSS, constants.crosstau)
```

## No decoherence.

```python
qd.crosstau = 0.
fss = np.linspace(-10, 10, 500)*1e-6
fidelities = [qd.ideal_fidelity_lorentzian(f)[0] for f in fss]
ghv = qd.ideal_fidelity_lorentzian(f)[1]

numerical_data_no_dec = np.loadtxt('out/fidelity_vs_fss/xtau1/fss_v_fidelity.txt', delimiter = ',')

plt.figure(figsize = (16./1.5,9./1.5))
plt.plot(numerical_data_no_dec[:,0], numerical_data_no_dec[:,1], 'r-', fss/1e-6, fidelities, 'b--')
plt.legend(['Monte carlo', 'Theoretical']) 
plt.text(6, 0.9, ' coherence ' + np.array(ghv).astype('|S3').tostring())
plt.xlim([-10, 10]) ; plt.ylim([0.45, 1.05])
plt.xlabel('Fine structure splitting $eV$') ; plt.ylabel('Fidelity')
```

## Decoherence time of 2.5 ns, coherence of 71%.

```python
qd.crosstau = 2.5
fss = np.linspace(-10, 10, 500)*1e-6
fidelities = [qd.ideal_fidelity_lorentzian(f)[0] for f in fss]
ghv = qd.ideal_fidelity_lorentzian(f)[1]

numerical_data_with_dec = np.loadtxt('out/fidelity_vs_fss/cross_dephasing2.5ns/fss_v_fidelity.txt', delimiter = ',')

plt.figure(figsize = (16./1.5,9./1.5))
plt.plot(numerical_data_with_dec[:,0], numerical_data_with_dec[:,1], 'r-', fss/1e-6, fidelities, 'b--')
plt.legend(['Monte carlo', 'Theoretical']) 
plt.text(6, 0.9, ' coherence ' + np.array(ghv).astype('|S4').tostring())
plt.xlim([-10, 10]) ; plt.ylim([0.45, 1.05])
plt.xlabel('Fine structure splitting $eV$') ; plt.ylabel('Fidelity')
```

## Polariation resolved corrolation curves with and without decoherence.

```python
def plot_corrs(arr11, arr12, arr21, arr22, arr31, arr32, title):
    delay = 180 - 20
    x_min = - 100
    x_max = 100
    
    fig = plt.figure(1, (12, 5))
    fig.subplots_adjust(left=0.05, right=0.98)
    
    fig.suptitle(title)
    
    grid = AxesGrid(fig, 111, nrows_ncols = (1, 3), axes_pad = 0.3, label_mode = "l")
    
    grid[0].plot(arr11[:,0]  - delay, arr11[:,1], 'b-', arr12[:,0] - delay, arr12[:,1], 'r-')
    grid[0].set_xlim([x_min, x_max]) ; grid[0].set_ylim([0, 2000])
    grid[0].set_yticks([]) ; grid[0].set_ylabel('Counts', fontsize = 14) ; grid[0].set_xlabel('$\\tau(ns)$', fontsize=14)
    grid[0].legend(['HH', 'HV'])
    
    grid[1].plot(arr21[:,0]  - delay, arr21[:,1], 'b-', arr22[:,0] - delay, arr22[:,1], 'r-')
    grid[1].set_xlim([x_min, x_max]) ; grid[1].set_ylim([0, arr21[:,1].max() + 10])
    grid[1].set_yticks([]) ; grid[1].set_xlabel('$\\tau(ns)$', fontsize=14)
    grid[1].legend(['DA', 'DD'])
    
    
    grid[2].plot(arr31[:,0]  - delay, arr31[:,1], 'b-', arr32[:,0] - delay, arr32[:,1], 'r-')
    grid[2].set_xlim([x_min, x_max]) ; grid[2].set_ylim([0, 200])
    grid[2].set_yticks([]) ; grid[2].set_xlabel('$\\tau(ns)$', fontsize=14)
    grid[2].legend(['LL', 'LR'])
```

```python
idx_fidelity = np.abs(np.array(numerical_data_no_dec[:,0] - 0)).argmin()
fidelity_no_dec = np.array(numerical_data_no_dec[idx_fidelity, 1]).astype('|S3').tostring()

plot_corrs(hh_no_dec_no_fss, hv_no_dec_no_fss, dd_no_dec_no_fss, da_no_dec_no_fss, ll_no_dec_no_fss, lr_no_dec_no_fss, 'No decoherence, fidelity: ' + fidelity_no_dec)
```
