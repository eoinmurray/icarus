# Degrees of corrolation in rectilinear, diagonal and circular basis versus cross dephasing lifetime.
No fine structure, no secondary emission, no background emission, no spin scattering.

```python
rootdir = '../out/2013-08-03/crosstau/'
dat = get_data(rootdir)

hold_crosstau = dat[0]
grect = dat[1][0]
gdiag = dat[1][1]
gcirc = dat[1][2]


plt.figure(figsize = (16/1.5, 9/1.5))

plt.plot(hold_crosstau, grect)
plt.plot(hold_crosstau, gdiag)
plt.plot(hold_crosstau, - gcirc)

plt.xlabel('$\\tau_{HV} (ns)$') ; plt.ylabel('Degree of corrolation')

plt.ylim([0,1.1])

plt.legend(['Rect', 'Diag', '- Circ'])
```

```python
plt.figure(figsize = (16/1.5, 9/1.5))
fidelities = (1 + grect + gdiag - gcirc)/4
plt.plot(hold_crosstau, - gcirc / grect)
plt.plot(hold_crosstau, gdiag / grect)

plt.xlabel('$\\tau_{HV} (ns)$') 
plt.ylim([0,1.1])
plt.legend([' - gcirc/grect', 'gdiag/grect'])
```
